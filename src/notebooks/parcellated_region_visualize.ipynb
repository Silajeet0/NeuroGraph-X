```python
import torch
import matplotlib.pyplot as plt

# 1. Path to ONE of your original 3D tensor regions
# (Make sure to use CNN_Tensors, not the 1D GNN_Features!)
tensor_path = "./data/CNN_Tensors/002_S_1155_I843510/Hippocampus_L.pt"

# 2. Load the tensor
# Shape should be (Depth, Height, Width)
region_3d = torch.load(tensor_path)
print(f"Loaded 3D Tensor Shape: {region_3d.shape}")

# 3. Find the middle index of each dimension
# (We use // 2 to get integer division)
mid_z = region_3d.shape[0] // 2  # Middle of Depth
mid_y = region_3d.shape[1] // 2  # Middle of Height
mid_x = region_3d.shape[2] // 2  # Middle of Width

# 4. Extract the 2D slices
# We slice along one axis and take all data (:) from the other two
slice_axial = region_3d[mid_z, :, :]
slice_coronal = region_3d[:, mid_y, :]
slice_sagittal = region_3d[:, :, mid_x]

# 5. Plot them side-by-side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Use the 'gray' colormap because MRI scans are grayscale
axes[0].imshow(slice_axial, cmap='gray')
axes[0].set_title(f"Axial Slice (Z={mid_z})")
axes[0].axis('off')

axes[1].imshow(slice_coronal, cmap='gray')
axes[1].set_title(f"Coronal Slice (Y={mid_y})")
axes[1].axis('off')

axes[2].imshow(slice_sagittal, cmap='gray')
axes[2].set_title(f"Sagittal Slice (X={mid_x})")
axes[2].axis('off')

plt.suptitle(f"Visualizing Left Hippocampus", fontsize=16)
plt.tight_layout()
plt.show()
```
